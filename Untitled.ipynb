```python
# Bringing in tensorflow
import tensorflow as tf
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus: 
    tf.config.experimental.set_memory_growth(gpu, True)
```

```python
from matplotlib import pyplot as plt
from tensorflow import keras
```

```python
batch_size = 16
image_height = 224
image_width = 224
data_dir = '../archive/cropped/train'

seed = 7
```

```python
training_ds = keras.utils.image_dataset_from_directory(
  data_dir,
  seed=seed,
  validation_split = 0.2,
  subset = "training",
  image_size = (image_height, image_width),
  batch_size=batch_size)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

training_ds = training_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
```

```python
# Do some data transformation
import numpy as np
```

```python
# Setup connection aka iterator
dataiterator = training_ds.as_numpy_iterator()
# Getting data out of the pipeline
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx in range(4):
    sample, labels = dataiterator.next()
    ax[idx].imshow(sample[idx][idx].astype("uint8"))
    ax[idx].title.set_text(labels[idx])
```
